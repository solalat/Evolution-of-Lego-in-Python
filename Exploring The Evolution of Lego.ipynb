# Project: Exploring The Evolution of Lego

Lego is a household name across the world, supported by a diverse toy line, hit movies, and a series of successful video games. In this project, I am going to explore a key development in the history of Lego: the introduction of licensed sets such as Star Wars, Super Heroes, and Harry Potter.
It may not be widely known, but Lego has had its share of ups and downs since its inception in the early 20th century. This includes a particularly rough period in the late 90s.
Lego was only able to survive due to a successful internal brand (Bionicle) and the introduction of its first licensed series: Star Wars. To complete this project, I will be answering few important questions

* set_num: A code that is unique to each set in the dataset. This column is critical, and a missing value indicates the set is a duplicate or invalid!.
* set_name: A name for every set in the dataset (note that this can be the same for different sets).
* year: The date the set was released.
* num_parts: The number of parts contained in the set. This column is not central to our analyses, so missing values are acceptable.
* theme_name: The name of the sub-theme of the set.
* parent_theme: The name of the parent theme the set belongs to. Matches the `name` column of the `parent_themes` csv file.






###### Import Libraries

```python
import pandas as pd
import numpy as np
```

###### Load Datasets

```python
# Load datasets
lego_sets = pd.read_csv('lego_sets.csv')
theme = pd.read_csv('parent_themes.csv')
```

### Column parent_theme in legos_sets and name in theme dataframes are the same hence the 2 datatsets will be merged

```python
merged_df = lego_sets.merge(theme, left_on = 'parent_theme', right_on = 'name')
merged_df.drop(columns = 'name_y', inplace=True)                                       #dropping the name_y column
merged_df.rename(columns = {'name_x':'theme_name'}, inplace = True)                   #renaming the name_x column
merged_df
```

```python
# check for null values in the column set_num
merged_df[merged_df['set_num'].isnull()].shape
```

```python
# drop the rows with null values
merged_df = merged_df.dropna(subset=['set_num'])
```

## What is the total number of parts per year?

```python
Num_Parts_Per_Year = merged_df.groupby('year').sum().reset_index()
Num_Parts_Per_Year = Num_Parts_Per_Year[['year', 'num_parts']].sort_values('num_parts', ascending = False)
Num_Parts_Per_Year
```

## How many sets were created in each century

```python
unique_year = merged_df['year'].unique()
unique_year.sort()
unique_year
```

```python
merged_df['century'] = np.where(merged_df['year'] > 2000, '21st century', '20th century')
merged_df
```

```python
# 
Set_Num_Century = merged_df.groupby('century').count().reset_index()
Set_Num_Century = Set_Num_Century[['century', 'set_num']]
Set_Num_Century
```

## What percentage of all licensed sets ever released were Star Wars themed?



```python
# retrieving only the licensed set from the is_licensed column where True
licensed = merged_df[merged_df['is_licensed']]
licensed.head()

```

```python
# retrieving only the Star Wars from the parent_theme column where parent_theme is equal to Star Wars
star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.head(10)
```

```python
# determining the percentage of licensed set
percentage_licensed_sets = star_wars.shape[0]/licensed.shape[0]
print(f"The percentage of licensed set is {percentage_licensed_sets:.0%}.")
```

## In which year was Star Wars not the most popular licenced theme (in terms of number of sets released that year)?

```python
licensed.head()
```

```python
# sorting the licensed dataframe by year and grouping by year and parent_theme 
df_sum = licensed.sort_values('year').groupby(['year', 'parent_theme']).sum().reset_index()
```

```python
max_df = df_sum.sort_values('is_licensed', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df
```

<b>In year 2017 Super Heroes replaced Star Wars as the most popular licensed theme</b>
